```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(2,9,10)
x
```

```python
y = x**2
y
```

```python
#representamos los valores de x respecto de y
```

```python
plt.plot(x,y,'red')
plt.title('Mi Grafico')
plt.xlabel('Eje X')
plt.ylabel('Eje Y')
plt.show()
```

```python
#se pueden a침adir varios graficos en el mismo espacio
```

```python
plt.subplot(1,2,1)  #(1,2,1) filas,columnas, posicion
plt.plot(x,y,'green')
plt.subplot(1,2,2)
plt.plot(y,x,'blue')
```

```python
plt.subplot(2,1,1)  
plt.plot(x,y,'green')
plt.subplot(2,1,2)
plt.plot(y,x,'blue')
```

#### Programaci칩n orienta a objetos  

vamos a hacer lo mismo pero desde la programaci칩n orienta a objetos

```python
figura =plt.figure()  #creamos una figura con el metdo plt.figure()
grafico = figura.add_axes([0.1,0.1,0.8,0.8])
grafico.plot(x,y,'green')
```

```python
figura =plt.figure()
grafico1 = figura.add_axes([0.1,0.1,0.8,0.8])  #add_axes([0.1,0.1,0.8,0.8]) son las separaciones y el tama침o
grafico1.plot(x,y,'r')

grafico2 = figura.add_axes([0.15,0.5,0.4,0.3])
grafico2.plot(y,x,'b')
```

## MULTIGRAFICOS

```python
figura, graficos = plt.subplots(nrows=2, ncols=2)
```

```python
#rellenamos los huecos dando posiciones a los gr치ficos
figura, graficos = plt.subplots(nrows=2, ncols=2)
graficos[0][0].plot(x,y,'b')
graficos[0][1].plot(x,y,'r')
graficos[1][0].plot(x,y,'g')
graficos[1][1].plot(x,y,'grey')

graficos[0][0].set_title('azul')
graficos[0][1].set_title('rojo')
graficos[1][0].set_title('verde')
graficos[1][1].set_title('gris')
```

#### Tama침o del gr치fico

```python
figura = plt.figure(figsize =(7,3))
grafico = figura.add_axes([0.1,0.1,0.9,0.9])
grafico.plot(x,y,'r')
```

### dos gr치ficos dentro de la misma figura

```python
figura = plt.figure(figsize =(7,3))
graficos = figura.add_axes([0.1,0.1,0.9,0.9])
graficos.plot(x,x**2,'r', label='Al Cuadrado')
graficos.plot(x,x**3,'g', label='Al Cubo')
#a침adimos la leyenda
graficos.legend(loc=0) #loc=0 es que busca la mejor posicion para la legend
```

#### Color del gr치fico,l칤neas y grosor

```python
figura = plt.figure(figsize =(7,3))
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y**2,'#a4ccee', label='Al Cuadrado', linewidth=10, alpha=0.5, linestyle='-.')
#los colores se pueden definir con colores hexadecimales
#linewidth=10, grosor de la linea
#alpha=0.5 es la transparencia
# linestyle estilo de la linea
grafico.legend(loc=0)
```

#### Marcadores

```python
#para marcar los puntos de cruce
```

```python
figura = plt.figure(figsize =(7,3))
grafico = figura.add_axes([0,0,1,1])
grafico.plot(x,y,'r', marker='*', markersize=8)
#https://matplotlib.org/stable/api/markers_api.html#module-matplotlib.markers
```
